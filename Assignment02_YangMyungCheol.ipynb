# Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.misc import derivative
```

# Define function $f(x) = x^2+3x+4$

```python
def func(x):
    result = x**2+3*x+4
    return result
```

# Define derivative of function

```python
def DeriFunc(x):
    return derivative(func, x)
```

# Define Taylor approximation

```python
def taylor_approx(domain, point):
    result = DeriFunc(point) * (domain - point) + func(point);
    return result
```

# Define domain and above function

```python
x = np.arange(-5,5,0.1)
y = func(x)
```

# Choose 3 points randomly

```python
point = np.zeros(3)

for i in range(3):
    point[i] = random.choice(x)
    print(point[i])
```

# Plot the funcation within the domatin

```python
plt.figure(1)
plt.plot(x, y, 'r', label="function")
plt.plot(x,taylor_approx(x,point[0]),'--c', label="taylor approximation on fist point")
plt.plot(x,taylor_approx(x,point[1]), ':g', label="taylor approximation on second point")
plt.plot(x,taylor_approx(x,point[2]), '-b', label="taylor approximation on third point")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

[Polynomial fitting]

Solve a least square problem to find an optimal polynomial curve for a given set of two dimensional points.

Demonstrate the effect of the degree of polynomial in fitting a given set of points.

- choose a polynomial curve and generate points along the curve with random noise

- plot the generated noisy points along with its original polynomial without noise

- plot the approximating polynomial curve obtained by solving a least square problem

- plot the approximating polynomial curve with varying polynomial degree

```python
%matplotlib inline
```

```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
import math
import random
```

### choose a polynomial curve and generate points along the curve with random noise


```python
np.random.seed(12)
x = np.linspace(0, 10, 20)

# polynomial curve : sin(x) + |x|*cos(x)
Origin_y = np.sin(x)+abs(x)*np.cos(x) 

# gnerate points along the curve with random noise
y = np.sin(x)+abs(x)*np.cos(x) + 2 * np.random.rand(20) * ( -1) ** (random.randint(0, 1))
```

### plot the generated noisy points along with its original polynomial without noise


```python
plt.plot(x, y, 'o', label = 'Noise')
plt.plot(x, Origin_y, '-', label = 'original graph')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("[Polynomial fitting]")

plt.legend()
plt.show()
```

### plot the approximating polynomial curve obtained by solving a least square problem


```python
def fit(x, y, n):
    arr = np.ones((x.shape[0], n + 1))
    for i in range (1, n + 1):
        arr.T[i] = x ** (i)
    a = np.asmatrix(arr)
    aT = np.transpose(a)
    t =inv(np.matrix(aT*a))*aT
    b = np.asmatrix(y)
    b1 = np.transpose(b)
    beta = t*b1
    yhat = np.dot(a,beta)
    return yhat
```

```python
fit2 = fit(x,y,2)
fit3 = fit(x,y,3)
fit4 = fit(x,y,4)
fit5 = fit(x,y,5)
fit6 = fit(x,y,6)
```

### plot the approximating polynomial curve with varying polynomial degree

```python
plt.plot(x, y, 'o', label = 'Noise')
plt.plot(x, Origin_y, '-', label = 'original graph')
#\plt.plot(x, np.sin(x), '-' , label = 'original function')
#plt.plot(x, yhat, '-' , label = 'poly = 2')
#plt.plot(x, y, '-' , label = 'original function')

plt.plot(x, fit2, '-' , label = 'poly = 2')


plt.xlabel("X")
plt.ylabel("Y")
plt.title("[Polynomial fitting]")

plt.legend()
plt.show()
```
